```python

import matplotlib.pyplot as plt
import numpy as np

vals = np.random.normal(0, 0.5, 10000)
plt.hist(vals, 50)
plt.show
```

```python
print(f"First moment (mean): {np.mean(vals)}")
print(f"Second moment (variance): {np.var(vals)}")
import scipy.stats as sp
print(f"3rd moment (skew): {sp.skew(vals)}")
print(f"4th moment (kurtosis): {sp.kurtosis(vals)}")
```
